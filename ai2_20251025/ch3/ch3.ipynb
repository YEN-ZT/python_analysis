````python
df = pd.read_csv('./tips.csv')
df.head()
````

````python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"
````

````python
df["Expensive"] = df["total_bill"].apply(yelp)
df.head()
````

````python
def quality(total_bill,tip):
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
````

````python
df["Tip Quality"] = df[["total_bill","tip"]].apply(lambda df:quality(df["total_bill"],df["tip"]),
                                                   axis=1)
````
